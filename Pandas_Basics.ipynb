```python
import pandas as pd
import requests as req
import base64
import numpy as np
```

```python
# parsing json directly

df = pd.read_json('https://restcountries.eu/rest/v2?fullText=true')
df
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.transpose.html

df = df.transpose()
df
```

```python
# dropping colums or rows

df = df.drop(['flag', 'nativeName', 'altSpellings', 'numericCode', 'translations'], axis=0)
df
```

```python
# renaming rows


df = df.rename(str.capitalize, axis='index')
df = df.rename({'Topleveldomain': 'TLD', 'Callingcodes': 'Calling_Codes', 'Latlng': 'Lat_Long',
               'Alpha2code': 'Alpha_2', 'Alpha3code': 'Alpha_3'})
df
```
