```python
def filter_fixtures(fixtures):
    team_1 = [i['scheduleEntry']['team1']['team']['fullName'] for i in fixtures['content']]
    team_2 = [i['scheduleEntry']['team2']['team']['fullName'] for i in fixtures['content']]
    city = [i['scheduleEntry']['venue']['city'] for i in fixtures['content']]
    venue = [i['scheduleEntry']['venue']['shortName'] for i in fixtures['content']]
    match_date = [i['scheduleEntry']['matchDate'].split('T')[0] for i in fixtures['content']]
    match_time_local = [i['scheduleEntry']['matchDate'].split('T')[1] for i in fixtures['content']]
    match_id = [i['scheduleEntry']['matchId']['id'] for i in fixtures['content']]
    tournament_id = [i['tournamentId']['id'] for i in fixtures['content']]
    begin_utc = [datetime.utcfromtimestamp(i['timestampMs']/1000).strftime('%H:%M') 
                 for i in fixtures['content']]
    end_utc = [datetime.utcfromtimestamp(i['endTimestampMs']/1000).strftime('%H:%M') 
                 for i in fixtures['content']]
    
    df = pd.DataFrame.from_dict(fixtures['content'])
    
    df = df.assign(Match_ID=match_id,
               Team_1=team_1, 
               Team_2=team_2, 
               City=city, 
               Venue=venue, 
               Match_Date=match_date, 
               Match_Time_Local=match_time_local,
               Tournament_ID=tournament_id, 
               Begin_UTC=begin_utc, 
               End_UTC=end_utc)

    df = df.drop(['resultOnly', 'timestamp', 'timestampMs', 'endTimestampMs'], axis=1)
    df = df[['Match_Date', 'Match_ID', 'Team_1','Team_2','label','tournamentLabel', 'Tournament_ID', 
        'Venue','City', 'Begin_UTC', 'End_UTC', 'Match_Time_Local']]
    return df
```

```python
url = 'https://cricketapi-icc.pulselive.com/fixtures/8199/scoring'
match_scorecard = req.get(url).json()
```

```python
def filter_innings(scorecard):
    innings = [i for i in scorecard['innings']]
    s = [i['scorecard'] for i in innings]
    runs = [i['runs'] for i in s]
    wkts = [i['wkts'] for i in s]
    all_out = [i['allOut'] for i in s]
    extras = [i['extras'] for i in s]
    declared = [i['declared'] for i in innings]
    over_progress = [i['overProgress'] for i in innings]
    run_rate = [i['runRate'] for i in innings]
    no_balls = [i['noBallRuns'] for i in extras]  
    leg_byes = [i['legByeRuns'] for i in extras] 
    byes = [i['byeRuns'] for i in extras] 
    wides = [i['wideRuns'] for i in extras] 
    
    df = pd.DataFrame([])
    
    df = df.assign(Runs=runs, 
                   Run_Rate=run_rate,
                   Wickets=wkts, 
                   All_Out=all_out, 
                   No_Balls=no_balls, 
                   Leg_Byes=leg_byes, 
                   Byes=byes, 
                   Wides=wides, 
                   Declared=declared, 
                   Over_Progress=over_progress)
    return df
```
